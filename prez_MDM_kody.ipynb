```python
data = pd.read_csv('data//PB_1996_2019_NH4.csv', sep=';')

data['Date'] = pd.to_datetime(data['Date'], format='%d.%m.%Y')

data_sorted = data.sort_values(by='Date')

# Funkcja do analizy sezonowoci dla danej stacji
def analyze_seasonality(station):
    station_data = data_sorted[data_sorted['ID_Station'] == station]
    station_data.set_index('Date', inplace=True)
    station_data = station_data.dropna()
    decomposition = seasonal_decompose(station_data['NH4'], model='additive', period=12)  # Okres = 12 miesicy
    plt.figure(figsize=(8, 6))
    plt.subplot(411)
    plt.plot(station_data.index, station_data['NH4'], label='Original')
    plt.legend(loc='best')
    plt.subplot(412)
    plt.plot(station_data.index, decomposition.trend, label='Trend')
    plt.legend(loc='best')
    plt.subplot(413)
    plt.plot(station_data.index, decomposition.seasonal,label='Seasonality')
    plt.legend(loc='best')
    plt.subplot(414)
    plt.plot(station_data.index, decomposition.resid, label='Residuals')
    plt.legend(loc='best')
    plt.tight_layout()

# Badanie sezonowoci dla stacji o ID 14
analyze_seasonality(14)
plt.show()
```

```python
from statsmodels.tsa.stattools import acf, pacf

# Wybr danych dla stacji o ID rwnym 14
station_14_data = data[data['ID_Station'] == 16]

# Konwersja kolumny 'Date' na typ daty
station_14_data['Date'] = pd.to_datetime(station_14_data['Date'], format='%d.%m.%Y')

# Ustawienie kolumny 'Date' jako indeks
station_14_data.set_index('Date', inplace=True)

# Obliczenie funkcji autokorelacji
lags = min(len(station_14_data['NH4']) // 2, 50)  # liczba opnie
acf_values = acf(station_14_data['NH4'], nlags=lags)
pacf_values = pacf(station_14_data['NH4'], nlags=lags)

# Wykresy funkcji autokorelacji
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
plt.stem(range(lags + 1), acf_values)
plt.xlabel('Lagi')
plt.ylabel('Autokorelacja')
plt.title('ACF dla stenia NH4 w stacji o ID rwnym 14')
plt.grid(color='gray', linestyle='-', linewidth=0.5)
plt.gca().set_facecolor('whitesmoke')

plt.subplot(1,2,2)
plt.stem(range(lags + 1), pacf_values)
plt.xlabel('Lagi')
plt.ylabel('Czciowa autokorelacja')
plt.title('PACF dla stenia NH4 w stacji o ID rwnym 14')
plt.grid(color='gray', linestyle='-', linewidth=0.5)
plt.gca().set_facecolor('whitesmoke')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Funkcja do przeprowadzenia testu ADF
def perform_adf_test(data):
    result = adfuller(data)
    print('Test Statistic:', result[0])
    print('P-value:', result[1])
    print('Lags Used:', result[2])
    print('Critical Values:', result[4])

    # Interpretacja wynikw
    if result[1] <= 0.05:
        print("Odrzucamy hipotez o niestacjonarnoci. Dane s stacjonarne.")
    else:
        print("Nie ma podstaw do odrzucenia hipotezy o niestacjonarnoci. Dane s niestacjonarne.")

# Test ADF dla danych ze stacji o ID rwnym 14
print("Test ADF dla stenia NH4 w stacji o ID rwnym 14:")
perform_adf_test(station_14_data['NH4'])
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Definicja maksymalnych wartoci p, d i q
max_p = 4
max_d = 4
max_q = 4

best_aic = float("inf")
best_order = None

# Iteracja po moliwych wartociach p, d i q
for p in range(max_p + 1):
    for d in range(max_d + 1):
        for q in range(max_q + 1):
            # Sprawdzenie, czy przynajmniej jedna z wartoci p, d lub q jest rna od zera
            if p != 0 or d != 0 or q != 0:
                try:
                    # Dopasowanie modelu ARIMA dla danej kombinacji p, d i q
                    model = ARIMA(station_14_data['NH4'], order=(p, d, q))
                    model_fit = model.fit()
                    # Sprawdzenie czy AIC jest nisze od aktualnie najlepszego
                    if model_fit.aic < best_aic:
                        best_aic = model_fit.aic
                        best_order = (p, d, q)
                except:
                    continue

print("Najlepszy model ARIMA:", best_order)
print("Warto AIC:", best_aic)
```

## 3. Usuwanie trendu i sezonowoci

```python
X = train_data['NH4']
t_list = np.arange(0,(len(X)),1)
a, b = np.polyfit(t_list, X, deg=1)
n = len(train_data)

X_no_trend = np.zeros(n)

for t in range(1,n):
    X_no_trend[t] =  X[t+2357] - a*t - b
    
plt.figure(figsize=(9, 5))
 
plt.scatter(t_list,X_no_trend, marker='o', linestyle='-', s=13)
plt.title("Stenie NH4 w stacji o ID rwnym 29 (bez trendu)")
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
def funcc(x,c,d,e):
    y = c*np.sin(d*x+e)
    return y

params, sth = curve_fit(funcc, t_list, X_no_trend, p0 = [3,0.004,0])
hat_c, hat_d, hat_e = params

X_no_seasonality = np.zeros(n)

for t in range(1,n):
    X_no_seasonality[t] =  X_no_trend[t] - hat_c*np.sin(hat_d*t+hat_e)

plt.figure(figsize=(9, 5))

plt.scatter(t_list, X_no_seasonality, marker='o', linestyle='-', s=13)
plt.title("Stenie NH4 w stacji o ID rwnym 29 (bez trendu i sezonowosci)")
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlabel('Kolejne numery dni pomiarowych') 
```

```python
p, q = 0, 2

# Wspczynniki modelu MA(2)
ma_params = [1, 0.5010, 0.1372]
sigma2 = 0.2189
```

```python
train_data_NH4 = train_data['NH4']

okres_sez = 30

# Liczba prbek i dugo danych
N = 10000
n = len(train_data[okres_sez:])

# Generowanie trajektorii ARMA
ar_models = [arma_generate_sample(ar=[1], ma=ma_params, nsample=n, scale=np.sqrt(sigma2) ) for _ in range(N)]
ar_seasonal_models = [ar_model + train_data_NH4[:-okres_sez] for ar_model in ar_models]

# Obliczanie kwantyli dla przedziau ufnoci
alpha = 0.05
quantiles_lower = np.quantile(ar_seasonal_models, alpha/2, axis=0)
quantiles_upper = np.quantile(ar_seasonal_models, 1 - alpha/2, axis=0) 

# Rysowanie wykresu
plt.figure(figsize=(12, 5)) 
line1, = plt.plot(range(n), quantiles_lower, c='orangered', linewidth=3/2, linestyle="dashdot") 
line2, = plt.plot(range(n), quantiles_upper, c='orangered', linewidth=3/2, linestyle="dashdot") 
line3 = plt.scatter(range(n), train_data_NH4[okres_sez:], s=30, c='blue', alpha=0.5) 
plt.legend([line1, line3], ['Przedziay ufnoci', 'Rzeczywiste wartoci danych']) 
plt.xlabel('Kolejne numery dni pomiarowych') 
plt.grid() 
plt.title("Porwnanie linii kwantylowych z trajektori danych")
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
# Obliczanie procentu danych w przedziale ufnoci
in_confidence_interval = np.logical_and(train_data_NH4[okres_sez:] >= quantiles_lower, train_data_NH4[okres_sez:] <= quantiles_upper)
percentage_in_confidence_interval = np.mean(in_confidence_interval) * 100

print("Procent danych w przedziale ufnoci: {:.2f}%".format(percentage_in_confidence_interval))
```
